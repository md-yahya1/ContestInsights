```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os
import time
```

```python
file = "../data/atcoder.csv"

refresh = True

if os.path.exists(file):

    last_modified = os.path.getmtime(file)

    hours_old = (time.time() - last_modified)/3600

    if hours_old < 48:

        print("Using cached data")

        df = pd.read_csv(file)

        refresh = False
```

```python
if refresh:

    print("Fetching fresh data...")

    url = "https://atcoder.jp/contests/"

    response = requests.get(url)

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    contests=[]

    rows=soup.select(
        "#contest-table-upcoming tbody tr"
    )

    for row in rows[:500]:

        cols=row.find_all("td")

        if len(cols)>=2:

            contests.append({
                "platform":"AtCoder",
                "date":cols[0].get_text(strip=True),
                "name":cols[1].get_text(strip=True)
            })

    df = pd.DataFrame(contests)

    df.to_csv(file,index=False)

    print("Saved")
```
