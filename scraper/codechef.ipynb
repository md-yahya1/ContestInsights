```python
import requests
import pandas as pd
import os
import time
```

```python
file = "../data/codechef.csv"

refresh = True

if os.path.exists(file):

    try:

        last_modified = os.path.getmtime(file)

        hours_old = (time.time()-last_modified)/3600

        if hours_old < 48:

            print("Using cached data")

            df = pd.read_csv(file)

            refresh = False

    except:

        print("Bad cache file, fetching again")
```

```python
if refresh:

    print("Fetching fresh data...")

    url = "https://www.codechef.com/api/list/contests/all"

    headers = {
        "User-Agent":"Mozilla/5.0"
    }

    response = requests.get(
        url,
        headers=headers,
        timeout=10
    )

    data = response.json()

    contests=[]

    for contest in data["future_contests"][:500]:

        contests.append({

            "platform":"CodeChef",

            "name":
            contest["contest_name"],

            "date":
            contest["contest_start_date"],

            "duration_mins":
            contest["contest_duration"]
        })

    df = pd.DataFrame(contests)

    df.to_csv(file,index=False)

    print("Saved")
```
