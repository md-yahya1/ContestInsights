```python
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python
file = "../data/codeforces.csv"

refresh = True

if os.path.exists(file):

    last_modified = os.path.getmtime(file)

    hours_old = (time.time() - last_modified) / 3600

    if hours_old < 48:

        print("Using cached data")

        df = pd.read_csv(file)

        refresh = False
```

```python
if refresh:

    print("Fetching fresh data...")

    url = "https://codeforces.com/api/contest.list"

    response = requests.get(url)

    data = response.json()

    contests = []

    for contest in data["result"][:100]:

        start_time = None

        if "startTimeSeconds" in contest:

            start_time = datetime.fromtimestamp(
                contest["startTimeSeconds"]
            )

        contests.append({
            "platform": "Codeforces",
            "name": contest["name"],
            "date": start_time,
            "phase": contest["phase"],
            "duration_mins":
                contest.get("durationSeconds",0)//60
        })

    df = pd.DataFrame(contests)

    df.to_csv(file, index=False)

    print("Data refreshed")
```
