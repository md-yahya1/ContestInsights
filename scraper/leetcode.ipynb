```python
import requests
import pandas as pd
import os
from datetime import datetime, timedelta
```

```python
file = "../data/leetcode.csv"

refresh = True

if os.path.exists(file):

    modified = datetime.fromtimestamp(
        os.path.getmtime(file)
    )

    if datetime.now() - modified < timedelta(hours=48):

        print("Using cached file...")

        df = pd.read_csv(file)

        refresh = False
```

```python
if refresh:

    url = "https://leetcode.com/graphql"

    query = """
    query {
      allContests {
        title
        startTime
        duration
      }
    }
    """

    response = requests.post(
        url,
        json={"query": query},
        headers={
            "User-Agent":"Mozilla/5.0"
        }
    )

    print(response.status_code)
```

```python
response.text[:500]
```

```python
data = response.json()

contests = []

for contest in data["data"]["allContests"]:

    contests.append({

        "platform":"LeetCode",

        "contest_name":
            contest["title"],

        "start_date":
            datetime.fromtimestamp(
                contest["startTime"]
            ),

        "duration_min":
            contest["duration"]/60
    })

df = pd.DataFrame(contests)

df.head()
```

```python
os.makedirs("../data", exist_ok=True)

df.to_csv(
    file,
    index=False
)

print("Saved", len(df), "rows")
```
